- https://doc.lsst.fink-broker.org/services/data_transfer/#diaobjectid-casting-errors

```python
import pyarrow.parquet as pq
```

```python
path_alerts = "ftransfer_lsst_2026-05-06_481298"
```

```python
# Read as a table
table = pq.read_table(path_alerts)
```

```python
# Make a list of diaObjectId
diaobjectids = [record["diaObjectId"] for record in table["diaObject"].to_pylist()]

# Tranform to Pandas keeping PyArrow types to prevent data corruption
import pandas as pd

df = table.to_pandas(types_mapper=pd.ArrowDtype)
```
