```python
import pandas as pd
from pathlib import Path
```

```python
# Dossier contenant les fichiers parquet
data_dir = Path("~/DATA/Fink2026/ftransfer_lsst_2026-04-07_771211").expanduser()

# Liste des fichiers
files = sorted(data_dir.glob("*.parquet"))
```

```python
files
```

```python
# Lecture + concaténation
df = pd.concat([pd.read_parquet(f) for f in files], ignore_index=True)

print(df.shape)
print(df.head())
```

```python
list(df.columns)
```

### ✅ 3. Si tu veux toutes les infos des sources (pas juste l’ID)

```python
# Par exemple récupérer (jd, mag, err, band) par objet :

cols = ["midpointMjdTai", "scienceFlux", "templateFlux", "snr"]

grouped = df.groupby("diaObjectId")[cols].apply(lambda x: x.to_dict("records"))

# 👉 Là tu obtiens pour chaque objet une liste de dicts = light curve brute
```
